```python
import os
import pandas as pd
from agents import *

df = pd.read_csv('data/example/example_data.csv')
accessions = df['Accession'].head(1)

for accession in accessions:
    print(f'Beginning Classification for {accession}')
    papers = os.listdir(f'data/library/{accession}')
    thermal_high_confidence = False  # Control flag

    with open(f'data/library/{accession}/{accession}.json', 'r') as f:
        metadata = json.load(f)

    # --- HOST CLASSIFICATION ---
    host_high_confidence = False
    best_host = None

    for paper in papers:
        print(f'Beginning Host Classification for {accession}/{paper}')
        with open(f'data/library/{accession}/{paper}') as f:
            paper_text = f.read()

        state = {
        'paper_text': paper_text,
        'source_path': f'data/library/{accession}/{paper}',
        'model': 'qwen3.5',
        'metadata': metadata,
        'host_species': None,
        'reasoning': None,
        'confidence': None,
        'thermal_range': None,
        'temperature': None,
        'thermal_reasoning': None,
        'thermal_confidence': None,
    }

        host_classification = IdentifyHost(state)

        if host_classification.get('host_species') != 'unknown' and host_classification.get('confidence') == 'high':
            with open('data/example/classification.txt', 'a') as f:
                f.write(
                    f'Accession: {accession}\n'
                    f'Paper: {paper}\n'
                    f'Host: {host_classification["host_species"]}\n'
                    f'Reasoning: {host_classification["reasoning"]}\n\n'
                )
            host_high_confidence = True
            best_host = host_classification['host_species']
            reasoning = host_classification['reasoning']
            confidence = host_classification['confidence']
            print(best_host, reasoning, confidence)
            break

    if not host_high_confidence:
        with open('data/example/classification.txt', 'a') as f:
            f.write(f'Accession: {accession}\nNo high-confidence result found.\n\n')



    # --- THERMAL CLASSIFICATION ---
    thermal_high_confidence = False

    for paper in papers:
        print(f'Beginning Thermal Classification for {accession}/{paper}')

        with open(f'data/library/{accession}/{paper}') as f:
            paper_text = f.read()

        state = {
        'paper_text': paper_text,
        'source_path': f'data/library/{accession}/{paper}',
        'model': 'qwen3.5',
        'metadata': metadata,
        'host_species': best_host, # set the host species
        'reasoning': reasoning,
        'confidence': confidence,
        'thermal_range': None,
        'temperature': None,
        'thermal_reasoning': None,
        'thermal_confidence': None,
    }

        thermal_classification = IdentifyThermalRange(state)

        if thermal_classification.get('thermal_range') != 'unknown' and thermal_classification.get('confidence') == 'high':
            with open('data/example/thermal_classification.txt', 'a') as f:
                f.write(
                    f'Accession: {accession}\n'
                    f'Paper: {paper}\n'
                    f'Thermal Range: {thermal_classification["thermal_range"]}\n'
                    f'Reasoning: {thermal_classification["thermal_reasoning"]}\n\n'
                )
            thermal_high_confidence = True
            state['thermal_range'] = thermal_classification['thermal_range']
            state['temperature'] = thermal_classification['temperature']
            state['thermal_reasoning'] = thermal_classification['thermal_reasoning']
            state['thermal_confidence'] = thermal_classification['thermal_confidence']
            break

    if not thermal_high_confidence:
        with open('data/example/thermal_classification.txt', 'a') as f:
            f.write(f'Accession: {accession}\nNo high-confidence result found.\n\n')

    print(state)
```
