```python

import pandas as pd
from tmp_software.tools import ConfusionMatrix
import matplotlib.pyplot as plt



actual=pd.read_csv('../example_data/expanded_example_data.csv')['Thermal Range']
predicted=pd.read_csv('fast/fast_result.csv')['thermal_range']
```

```python
#Clean out None
actual_clean = actual.fillna("None")
pred_clean = predicted.fillna("None")

valid = ['mesophile', 'psychrophile', 'thermophile', 'None']

# keep only rows where BOTH actual and predicted are valid labels

mask = actual_clean.isin(valid) & pred_clean.isin(valid)

actual_filtered = actual_clean[mask]

pred_filtered = pred_clean[mask]

plot = ConfusionMatrix(actual_filtered, pred_filtered)

plot.plot(cmap='Blues')

plt.savefig('fast/fast_confusion_matrix_clean.png', dpi=300, bbox_inches='tight')

```
