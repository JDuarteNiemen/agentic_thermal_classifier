```python
from langgraph.graph import StateGraph, END
from typing import TypedDict
from IPython.display import Image

# Define your state shape
class State(TypedDict):
    message: str

# Define your node functions
def node_a(state: State) -> State:
    return {"message": "hello from A"}

def node_b(state: State) -> State:
    return {"message": "hello from B"}

def node_c(state: State) -> State:
    return {"message": "hello from C"}

def node_d(state: State) -> State:
    return {"message": "hello from D"}

def node_e(state: State) -> State:
    return {"message": "hello from E"}

def node_f(state: State) -> State:
    return {"message": "hello from F"}


# Build the graph
graph = StateGraph(State)

graph.add_node("EXTRACT METADATA", node_a)
graph.add_node("Write useful metadata to csv", node_b)
graph.add_node("Extract papers", node_c)
graph.add_node("Parse papers for information on host organism", node_d)
graph.add_node("Literature search onm host organism for information about optimal growth conditions", node_e)

# Add edges
graph.set_entry_point("EXTRACT METADATA")
graph.add_edge("EXTRACT METADATA", "Write useful metadata to csv")
graph.add_edge("EXTRACT METADATA", "Extract papers")
graph.add_edge("Extract papers", "Parse papers for information on host organism")
graph.add_edge("Parse papers for information on host organism", "Literature search onm host organism for information about optimal growth conditions")
graph.add_edge("Write useful metadata to csv", END)
graph.add_edge("Literature search on host organism for information about optimal growth conditions", END)


# Compile and visualise
app = graph.compile()
Image(app.get_graph().draw_mermaid_png())
```
