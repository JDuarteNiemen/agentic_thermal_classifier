```python
# IDENTIFYING THERMAL RANGE USING ONLY THE NAME OF THE PHAGE
from no_context import *
from tools import *
import pandas as pd
import time

df=pd.read_csv('data/example/example_data.csv')
accessions=df['Accession']

with open(f'results/no_context/thermal_name_classification.txt','w') as f:
    f.write(f'Classifying thermal range of bacteriophage using only the name of the phage\n')
    f.write(f'Forced output, must return a classification\n')

with open(f'results/no_context/thermal_name_classification.csv','w') as f:
    f.write('accession,thermal_range,duration(m)\n')
full_start=time.time()
for accession in accessions:
    with open(f'results/no_context/thermal_name_classification.txt','a') as f:
        f.write(f'============={accession}=============\n')
    metadata=FetchNcbiMetadata(accession)


    state = {
            'paper_text': None,
            'source_path': None,
            'model': 'qwen3.5',
            'metadata': None,
            'phage': metadata.get('organism', 'unknown'),
            'host_species': None,
            'reasoning': None,
            'confidence': None,
            'thermal_range': None,
            'temperature': None,
            'thermal_reasoning': None,
            'thermal_confidence': None,
        }
    acc_start=time.time()
    out=IdentifyThermalRangeName(state)
    acc_end=time.time()
    acc_duration=(acc_end-acc_start)/60
    print(f'{accession}: {out.get("thermal_range")}')


    with open(f'results/no_context/thermal_name_classification.csv','a') as f:
        f.write(f'{accession},{out.get("thermal_range", "unknown")},{round(acc_duration,2)}\n')

    with open(f'results/no_context/thermal_name_classification.txt','a') as f:
        f.write(f'Thermal Classification: {out.get("thermal_range", "unknown")}\n')
        f.write(f'Duration: {round(acc_duration,2)}\n')

full_end=time.time()
full_duration=(full_end-full_start)/60
with open(f'results/no_context/thermal_name_classification.txt','a') as f:
    f.write(f'\n\nFull Duration: {round(full_duration,2)} minutes\n')
```
