```python
###########################################################################################################################################################
# QWEN3.5
###########################################################################################################################################################


# IDENTIFYING THE THERMAL RANGE OF PHAGE FROM METADATA
from no_context import *
from tools import *
import pandas as pd
import time
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

df=pd.read_csv('data/example/example_data.csv')
accessions=df['Accession']

with open(f'results/no_context/qwen3.5/thermal_metadata_classification.txt','w') as f:
    f.write(f'Classifying thermal range of bacteriophage using only the metadata\n')
    f.write(f'Forced output, must return a classification\n')

with open(f'results/no_context/qwen3.5/thermal_metadata_classification.csv','w') as f:
    f.write('accession,thermal_range,duration(m)\n')
full_start=time.time()
for accession in accessions:
    with open(f'results/no_context/qwen3.5/thermal_metadata_classification.txt','a') as f:
        f.write(f'============={accession}=============\n')
    metadata=FetchNcbiMetadata(accession)

    state = {
            'paper_text': None,
            'source_path': None,
            'model': 'qwen3.5',
            'metadata': metadata,
            'phage': metadata.get('organism', 'unknown'),
            'host_species': None,
            'reasoning': None,
            'confidence': None,
            'thermal_range': None,
            'temperature': None,
            'thermal_reasoning': None,
            'thermal_confidence': None,
        }
    acc_start=time.time()
    out=IdentifyThermalRangeNoContext(state)
    acc_end=time.time()
    acc_duration=(round(acc_end-acc_start,2)/60)
    print(f'{accession}: {out.get("thermal_range")}')


    with open(f'results/no_context/qwen3.5/thermal_metadata_classification.csv','a') as f:
        f.write(f'{accession},{out.get("thermal_range", "unknown")},{acc_duration}\n')

    with open(f'results/no_context/qwen3.5/thermal_metadata_classification.txt','a') as f:
        f.write(f'Thermal Classification: {out.get("thermal_range", "unknown")}\n')
        f.write(f'Duration: {acc_duration}\n')

full_end=time.time()
full_duration=(round(full_end-full_start,2)/60)
with open(f'results/no_context/qwen3.5/thermal_metadata_classification.txt','a') as f:
    f.write(f'\n\nFull Duration: {full_duration} minutes\n')

pred_df=pd.read_csv('results/no_context/qwen3.5/thermal_metadata_classification.csv')

# Create confusion matrix
actual=df['Thermal Range']
pred=pred_df['thermal_range']
plot=ConfusionMatrix(actual, pred)
plot.plot(cmap='Blues')
plt.savefig('results/no_context/qwen3.5/confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()


# Get the classification report
classification_report=classification_report(actual, pred)

with open('results/no_context/qwen3.5/classification_report.txt', 'w') as f:
    f.write(classification_report)
```
